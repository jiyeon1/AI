# K-Means Clustering 모델

[k-means algorithm
](https://process-mining.tistory.com/122)
- 데이터의 어떤 영역을 대표하는 클러스터 중심을 찾음
- 서로 유사한 데이터들은 같은 그룹으로, 서로 유사하지 않은 데이터는 다른 그룹으로 분리하는 것이 목표

1. k개의 데이터 포인트를 가장 가까운 클러스터 영역에 할당
2. 클러스터에 할당된 데이터 포인트의 평균으로 클러스터 중심을 다시 지정
3. 클러스터에 할당되는 데이터 포인트에 변화 없료 때 알고리즘 종료

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
import numpy as np
```

## 가상의 데이터 생성
* 분류용 데이터 생성 메서드를 이용해서 label이 없는 데이터를 생성하였다. 
- n_samples : 표본 데이터의 수
- n_features : 독립변수의 수
- centers : 클러스터 수, or 중심
- cluster_std : 클러스터 표준편차
- random_state : 난수 발생 시드

- 반환값 : x(독립변수) [n_samples, n_features] 크기의 배열  
  y(종속변수) [n_samples] 크기의 배열  


```python
X, y = make_blobs(n_samples=100, centers=1, cluster_std=0.5, random_state=30)
# print(X)
X[:10]
```

```python
plt.scatter(X[:, 0], X[:, 1], c='blue', edgecolor='k', lw=0.5)  # (x축, y축, ..)
plt.show()
```

## KMeans (초기화 = random)

```python
model_kpp = KMeans(n_clusters=3, init="random").fit(X)  # 클러스터 개수 정하
```

```python
model_kpp.fit(X)  # max_iter 까지 반복해 centroid 정하
```

Attributes of sklearn.clustering.KMeans
    
- cluster__centers_ : Coordinates of cluster centers

- labels_ : Labels of each point

- inertia_ : Sum of squared distances of samples to their closest cluster center.

- n__iter_ : Number of iterations run.

```python
center = np.zeros([3, 2]) # 클러스터 중심을 넣을 배열 생성
center = model_kpp.cluster_centers_ # centroid 위치
center
```

```python
c0 = center[0]
c1 = center[1]
c2 = center[2]
```

```python
plt.scatter(X[model_kpp.labels_ == 0, 0],X[model_kpp.labels_ == 0, 1], marker='o', c='g', edgecolor='k', lw=0.5)
plt.scatter(X[model_kpp.labels_ == 1, 0],X[model_kpp.labels_ == 1, 1], marker='o', c='r', edgecolor='k', lw=0.5)
plt.scatter(X[model_kpp.labels_ == 2, 0],X[model_kpp.labels_ == 2, 1], marker='o', c='b', edgecolor='k', lw=0.5)
plt.scatter(c0[0], c0[1], s=200, c="g", edgecolor='k', lw=1)
plt.scatter(c1[0], c1[1], s=200, c="r", edgecolor='k', lw=1)
plt.scatter(c2[0], c2[1], s=200, c="b", edgecolor='k', lw=1)
plt.show()
```

```python
import tensorflow as tf
tf.__version__
```

#### tensor
numpy에 ndarray라면, tensorflow에서는 tensor다.  
data 의 기본 단위가 tensor라고 생각하자.  
구조는 ndarray와 동일하다. 즉, 배열 형태이다.  
numpy shape = 몇 개의 데이터, 몇 차원   

```python
var = tf.Variable(5)
var
```

```python
con = tf.constant(5)
con
```

```python
add = var + con #tf.add(var, con)
add
```

var 와 con의 차이는 뭘까?  
var는 변형가능, con은 변형 불가능하다.  
무슨 말일까? 한번 살펴보자.

```python
init = tf.global_variables_initializer()
```

```python
assign = tf.assign(var, 10) #var <- 10 변형
```

assign = tf.assign(con, 10) #con <- 10 변형

#### session : build 따로 run 따로

var에 5를 넣고, var을 출력해봤더니, 5가 안나온다.  
build와 run을 구분!

##### tensor들의 graph형태로 이루어져 있다. 

```python
sess = tf.Session() #스위치 공사
sess.run(init)

print('con:', sess.run(con))  # con = 5
print('var:', sess.run(var))  # var = 5

print(sess.run(add))    # add = 5

print(sess.run(assign))   # var = 10

print('after assign, var:', sess.run(var), '// add : ', sess.run(add))
sess.close()
```

```python
with tf.Session() as sess:  # with : close 명시 안해도 자동으로 종료
    sess.run(init)

    print('con:', sess.run(con))
    print('var:', sess.run(var))

    print(sess.run(add))

    print(sess.run(assign))

    print('after assign, var:', sess.run(var))
```
